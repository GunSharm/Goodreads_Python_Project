# Find Duplicates in titles

```python
gd['title'].duplicated().sum()
```

We have a lot of duplicates in the title column. Have to explore more what caused this duplication.

```python
dup_mask = gd['title'].duplicated()
gd.loc[dup_mask, 'title']
```

```python
gd.loc[gd['title'] == 'Collected Poems', :]
```

For 'Collected Poems', the title is used for different books with the same generic name to name a book of collection of poems. So we have to alter it to include the authors' names so as to differentiate it. 

```python
gd.loc[gd['title'] == 'Limits of Destiny', :]
```

Some books have the same name as they belong to the same book series.

```python
gd.loc[gd['title'] == 'Twisted', :]
```

For 'Twisted' we see that different books can have the same names belonging to different series or standalone books.

So in conclusion, we based on our exploration, we conclude that duplication of titles is okay as it is common.
